```python
frames = mic.record_x_seconds(seconds=10)
```

```python
with wave.open('./test_nat.wav', 'wb') as wave_file:
           wave_file.setnchannels(MIC_NUMBER_OF_CHANNELS)
           wave_file.setsampwidth(4)
           wave_file.setframerate(MIC_RATE)
           wave_file.writeframes(b''.join(frames))
```

```python
y, sr = soundfile.read('./test_nat.wav')
log_s = extract_spectrogram(y,sr)
#new plot
fig = plt.figure(figsize=(12,4))
ax = plt.Axes(fig, [0., 0., 1., 1.])
ax.set_axis_off()
fig.add_axes(ax)

#getting spectrogram
specdisplay.specshow(log_s, sr=sr, x_axis='time', y_axis='mel')
plt.show()
```
